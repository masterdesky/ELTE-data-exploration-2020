## Exercise 3

How many deaths were within a given radius around the wells? Test for a series of radiuses. Project your points into a coordinate system where you can measure real distances (using e.g. `pyproj`), then create a meaningul visualization of the data.

*It is crucial to use the appropriate coordinate system for your purposes. For this exercise, the spherical coordinate system of lon,lat values is not convenient, since we cannot calculate distances in meters as easily as we would in a Cartesian coordinate system. Because of the Earth's shape, most projections distort distances at certain latitudes and longitudes.*

*In GIS, different coordinate systems are called CRS (Coordinate Reference System), and have patented codes that define the origin, the projection etc. The usual lon,lat pairs are called the WGS84 system. A good choice for a Cartesian system is for example the Pseudo-Mercator projection. The `pyproj` library is especially lightweight to use in Python.*

```python
deathsProj = deaths["geometry"].to_crs({'init': 'epsg:3395'})
pumpsProj = pumps["geometry"].to_crs({'init': 'epsg:3395'})
```

```python
from functools import reduce
distancesFromWells = pumpsProj.map (lambda x: deathsProj.distance(x))
```

## Exercise 5

Calculate the Voronoi cells for the pumps, and count the number of deaths falling into each of the cells.  Create a bar chart of the number of deaths in each cell. Pinpoint the possible source of the infection.

*Store the Voronoi cells in a GeoDataFrame, then you can easily do a spatial join (e.g. testing whether a polygon in the Voronoi dataframe contains a point from the death dataframe).*

```python
from scipy.spatial import Voronoi
from shapely.geometry import MultiPoint
from shapely.geometry import Polygon, box

wellCoords = [x.coords[0] for x in pumpsProj]

def AddOutOfBoundEdgeCoords (coords):
    bounds = MultiPoint(coords).bounds
    coords.append ([bounds[0] - 10000, bounds[1] - 10000])
    coords.append ([bounds[0] - 10000, bounds[3] + 10000])
    coords.append ([bounds[2] + 10000, bounds[1] - 10000])
    coords.append ([bounds[2] + 10000, bounds[3] + 10000])
    
    global bufferedBoundingBox
    bufferedBoundingBox = box (*bounds).buffer(100)

AddOutOfBoundEdgeCoords (wellCoords)

vor = Voronoi(wellCoords)

regions = [[vor.vertices[y] for y in x] for x in vor.regions if (-1 not in x) and x != []]
regionIndices = np.arange(len(vor.regions))[[-1 not in x for x in vor.regions]][1:]
```

```python
polyFrame = gpd.GeoDataFrame(crs={'init': 'epsg:3395'}, geometry=[Polygon(hull).intersection (bufferedBoundingBox) for hull in regions])
polyFrame["wellID"] = [ np.where(vor.point_region==y)[0][0]+1 for y in [regionIndices[x] for x in np.arange(len(regions))]]
polyFrame4326 = polyFrame.to_crs({'init': 'epsg:4326'})
```

```python
polyFrame.head()
```

```python
pointPolyJoint = gpd.sjoin (polyFrame4326, deaths[["geometry", "OBJECTID"]], op="contains")
pointPolyJoint.head()
```

```python
tempIndex, counts = numpy.unique(pointPolyJoint.wellID, return_counts=True)
plt.bar(tempIndex, counts)
plt.xticks(np.arange(1, len(pumps)+1))
plt.xlim (0, len(pumps)+1)
plt.xlabel("pump index")
plt.ylabel("Deaths closest to pump")

deathsByWell = [0] * len(pumpsProj)
for ind in range(len(tempIndex)):
    deathsByWell[tempIndex[ind]-1] = counts[ind]

badIndex = tempIndex[np.argmax(counts)]
```

```python
def geoJSONStyle(feature):
    return {'color': "red" if feature['properties']['wellID'] == badIndex else "blue"}

polyMap = folium.Map(location=[51.5134, -0.136],zoom_start=16, tiles='cartodbpositron')
polyJSON = json.loads(polyFrame4326.to_json())
folium.GeoJson(polyJSON, style_function=geoJSONStyle).add_to(polyMap)

for ind, pump in pumps.iterrows():
    marker = folium.Marker(location=[pump.lat, pump.lon], icon=folium.Icon(icon="glyphicon-tint", color="red" if pump.OBJECTID == badIndex else "blue"), popup="pump " + str(pump.OBJECTID))
    marker.add_to(polyMap)

polyMap
```

```python
import scipy.stats as st

#a fuggveny fele erosen stackoverflow altal van ihletve
#tudom mit csinal, de nem fogom csak azert atnevezni a
#valtozokat, hogy ne legyen karakterpontosan ugyanaz
def MakeDensityPlotAndWriteWellDensities (data):
    global pumpDensities
    global infectedIndexDensity
    
    x = data[:, 0]
    y = data[:, 1]
    xmin, xmax = x.min()-200, x.max()+200
    ymin, ymax = y.min()-200, y.max()+200

    xx, yy = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]
    positions = np.vstack([xx.ravel(), yy.ravel()])
    values = np.vstack([x, y])
    kernel = st.gaussian_kde(values)
    f = np.reshape(kernel(positions).T, xx.shape)
        
    pumpDensities = []
    for pumpInd in range(len(pumpsProj)):
        px = np.digitize(pumpsProj[pumpInd].coords[0][0], np.arange(xmin, xmax, (xmax-xmin)/100))
        py = np.digitize(pumpsProj[pumpInd].coords[0][1], np.arange(ymin, ymax, (ymax-ymin)/100))
        density = f[px, py] / 10000
        
        pumpDensities.append (density)
        print ("Pump %i: %0.2e %%/m2" % (pumpInd+1, density) )
    
    infectedIndexDensity = np.argmax(pumpDensities)
    
    fig = plt.figure(figsize=(10,10))
    ax = fig.gca()
    ax.set_xlim(xmin, xmax)
    ax.set_ylim(ymin, ymax)
    
    ax.imshow(np.rot90(f), cmap='Reds', extent=[xmin, xmax, ymin, ymax])
    
    # Contour plot
    cset = ax.contour(xx, yy, f, colors='k')
    ax.clabel(cset, inline=1, fontsize=10, fmt=r"%0.e")
    
    ax.set_xlabel('X [m]')
    ax.set_ylabel('Y [m]')
    
    pumpsProj[np.arange(len(pumpsProj)) != infectedIndexDensity].plot(ax=ax, markersize=200)
    pumpsProj[[infectedIndexDensity]].plot(ax=ax, color="yellow", markersize=200)

MakeDensityPlotAndWriteWellDensities (np.array(list(map(lambda x: np.asarray(x.coords[0]), deathsProj.values))))
```

*[GeoJSON](http://geojson.org/) is a common format that is used in web applications. It is basically a simple key-value dictionary with a predefined structure for storing geographic information. Once loaded into Python, it behaves like a normal Python dictionary, thus creating or modifying it is quite straightforward. When reading or writing to a file, the `json` module is handy. The advice is to write a function that converts a list of coordinates from a polygon into a GeoJSON feature, then add it to your GeoJSON structure.*

```python
import xml.etree.ElementTree as ET
tree = ET.fromstring(response.content)
```

```python
def AddNode (node):
    for item in node.items ():
        if item[0] == "id":
            id = item[1]
        elif item[0] == "lat":
            lat = item[1]
        elif item[0] == "lon":
            lon = item[1]
    
    nodes[id] = (float(lat), float(lon))
```

```python
def AddWay (way):
    myNodes = []
    isStreet = False
    
    for tag in way.getchildren ():
        if tag.tag == "nd":
            for item in tag.items ():
                if item[0] == "ref":
                    myNodes.append (item[1])
        elif tag.tag == "tag":
            for item in tag.items ():
                if item[0] == "k" and item[1] == "highway":
                    isStreet = True
    
    if isStreet:
        ways.append(myNodes)
```

```python
nodes = {}
ways = []

for xmlNode in tree.getchildren():
    if xmlNode.tag == "node":
        AddNode (xmlNode)
    elif xmlNode.tag == "way":
        AddWay (xmlNode)
```
